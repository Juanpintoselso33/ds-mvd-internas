```python
import numpy as np
import matplotlib.pyplot as plt

# Datos originales
listas = ['2250', '50', '500', '300', '1120', '44', '2007', '60', '75', '11', '350', '2500', '222', '55', '34', '1204', '21250', '32', '307', '251', '104', '330', '36', '289', '161']
votos = [2276, 1457, 497, 261, 213, 200, 193, 189, 132, 102, 97, 86, 86, 84, 83, 56, 54, 51, 48, 35, 34, 33, 30, 12, 2]

# Sort data by vote count
sorted_indices = np.argsort(votos)[::-1]
listas = np.array(listas)[sorted_indices]
votos = np.array(votos)[sorted_indices]

# Group smaller slices into "Others"
threshold = 0.015  # 1.5% threshold
total_votes = sum(votos)
others_mask = (votos / total_votes) < threshold
others_votes = np.sum(votos[others_mask])
listas = np.append(listas[~others_mask], 'Otras')
votos = np.append(votos[~others_mask], others_votes)

# Calculate percentages
porcentajes = 100 * votos / total_votes

# Custom color palette based on the logo
base_colors = ['#1E90FF', '#FFD700', '#000080']  # Light blue, gold, navy blue
n_colors = len(listas)

# Generate a color palette
colors = []
for i, lista in enumerate(listas):
    if lista == '50':
        colors.append('#1E90FF')  # Assign light blue to '50'
    elif lista == '2250':
        colors.append('#4169E1')  # Assign a different blue to '2250'
    else:
        ratio = i / (n_colors - 1)
        if ratio < 0.5:
            c1, c2 = base_colors[1], base_colors[2]
        else:
            c1, c2 = base_colors[2], base_colors[1]
        t = (ratio - 0.5) * 2 if ratio >= 0.5 else ratio * 2
        
        r = int((1-t) * int(c1[1:3], 16) + t * int(c2[1:3], 16))
        g = int((1-t) * int(c1[3:5], 16) + t * int(c2[3:5], 16))
        b = int((1-t) * int(c1[5:7], 16) + t * int(c2[5:7], 16))
        
        colors.append(f'#{r:02x}{g:02x}{b:02x}')

# Create pie chart
fig, ax = plt.subplots(figsize=(14, 10))
explode = [0.05 if lista == '50' else 0 for lista in listas]
wedges, _ = ax.pie(votos, labels=None, colors=colors, startangle=90, 
                   wedgeprops=dict(width=0.5, edgecolor='white'),
                   explode=explode)

# Add labels with percentages
for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1) / 2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    
    if listas[i] == '50':
        fontsize = 14
        weight = 'bold'
        x *= 1.15  # Move label slightly further out
        y *= 1.15
    else:
        fontsize = 10
        weight = 'normal'
        x *= 1.1
        y *= 1.1
    
    ax.text(x, y, f"{listas[i]}\n{porcentajes[i]:.1f}%", 
        horizontalalignment=horizontalalignment, 
        verticalalignment='center', fontsize=fontsize, fontweight=weight)

# Title and aspect
ax.set(aspect="equal", title='Votos al ODN de Por la Patria en Montevideo')
ax.set_title('Votos al ODN de Por la Patria en Montevideo', fontsize=16, fontweight='bold')

# Remove axes
ax.axis('off')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
